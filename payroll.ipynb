 Employee Data in DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample employee data in a DataFrame
employee_data = {
    'Employee ID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Department': ['IT', 'HR', 'Finance'],
    'Hours Worked': [40, 35, 30],
    'Hourly Rate': [50, 45, 55],
    'Tax %': [10, 15, 12]
}

# Create DataFrame
df = pd.DataFrame(employee_data)
print("Initial Employee Data:")
print(df)
```

Payroll Calculation Using NumPy:

```python
# Function to calculate gross salary and net salary
def calculate_salaries(df):
    # Calculate gross salary
    df['Gross Salary'] = df['Hours Worked'] * df['Hourly Rate']

    # Calculate tax amounts
    df['Tax Amount'] = df['Gross Salary'] * (df['Tax %'] / 100)

    # Calculate net salary
    df['Net Salary'] = df['Gross Salary'] - df['Tax Amount']

    return df

# Apply the payroll calculation
df = calculate_salaries(df)
print("\nUpdated Employee Data with Salaries:")
print(df)
```

Save Payroll Data to Excel:

```python
df.to_excel("payroll_data.xlsx", index=False)
print("\nPayroll data saved to payroll_data.xlsx")
```
